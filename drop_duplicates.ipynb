```python
import pandas as pd
```

```python
data = pd.read_csv("ipl.csv")
```

```python
data.drop_duplicates("city") #Give unique occurance of each city
```

```python
data.drop_duplicates("season") #Give frst match of each season and drop all other
```

```python
data.drop_duplicates("season" , keep="last") #Give Last match of each season means final game
```

```python
data.drop_duplicates("season" , keep="last")[['season' , 'winner']].sort_values("season") #Give clean winner list of each season
```
