```python
import pandas as pd
```

```python
data = pd.read_csv("deliveries.csv")
```

```python
data
```

```python
batsman = data.groupby("batsman")
```

```python
batsman["batsman_runs"].sum().sort_values(ascending=False)
```

```python
mask = data["batsman_runs"] == 6
newData = data[mask]
newData.groupby("batsman").size().sort_values(ascending=False).head()
```
